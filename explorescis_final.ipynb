## Mapping the location of Hawker Centres onto the Map


## Introduction to Foilum

Folium is a powerful data visualization library in Python that was built primarily to help people visualize geospatial data. With Folium, one can create a map of any location in the world if its latitude and longitude values are known.

To understand the density of the listings spatially, we can use the folium heat map to plot the listings. There are 2 broad steps to creating a heatmap:

i) Creating a folium base map of the location <br>
ii) Adding the heatmap layer

## Importing the Folium library

```python
import folium
```

##Initialising a map

```python
map = folium.Map(location=[1.3437459, 103.8240449], zoom_start=12, control_scale=True)
```

```python
display(map)
```

## Adding the locations onto the map

```python
hawker_centre_locations = hawker_df[["latitude_hc", "longitude_hc", 'name']]
```

```python
for index, location_info in hawker_centre_locations.iterrows():
    folium.Marker([location_info["latitude_hc"], location_info["longitude_hc"]], popup=location_info["name"]).add_to(map)
```
